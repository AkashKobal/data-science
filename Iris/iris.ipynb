Splitting Dataset in X and Y

```python

# here columns ('sepal_length', 'sepal_width', 'petal_length', 'petal_width') will be X
# and column('species') will be Y

X = iris.iloc[:,:4] #first four coumlns
Y = iris.iloc[:,-1] #last column

```

```python
print(X)
```

```python
print(Y)
```

```python
# Data Normalization
x = preprocessing.StandardScaler().fit_transform(X)
print(x)
```

```python
# Train Test Split 
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size=0.2,random_state=0)
print(x_train.shape)
```

```python
# Data Model and Predicting 
knnmodel = KNeighborsClassifier(n_neighbors=3)
knnmodel.fit(x_train,y_train)
```

```python
y_predict = knnmodel.predict(x_test)
```

```python
#Accuracy Score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```
